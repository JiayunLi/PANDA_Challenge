```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file = "./input/csv_pkl_files/4_fold_whole_train.csv"
```

```python
df = pd.read_csv(csv_file)
```

```python
fig = plt.figure(figsize=(10,6))
ax = sns.countplot(x="isup_grade", hue="data_provider", data=df)
for p in ax.patches:
    '''
    Courtesy of Rohit Singh for teaching me this
    https://www.kaggle.com/rohitsingh9990/panda-eda-better-visualization-simple-baseline
    '''
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2,
                height +3,
                '{:1.2f}%'.format(100*height/10616),
                ha="center")
```

```python
for fold in range(4):
    df.head()
    df_fold = df.loc[df["split"] == fold]
    fig = plt.figure(figsize=(10,6))
    ax = sns.countplot(x="isup_grade", hue="data_provider", data=df_fold)
    for p in ax.patches:
        '''
        Courtesy of Rohit Singh for teaching me this
        https://www.kaggle.com/rohitsingh9990/panda-eda-better-visualization-simple-baseline
        '''
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2,
                    height +3,
                    '{:1.2f}%'.format(100*height/len(df_fold)),
                    ha="center")
```
